```python
# Make a function to copy files

def copy_files(data, destination_folder,
               source_folder = "/Users/daniil/Library/CloudStorage/OneDrive-IndianaUniversity/Documents/Experiments/IU/Memorability Rocks/Nosofsky files/Stimuli_Rocks480"):
    """
    Copy files from one folder to another
    
    Parameters:
        data: Pandas DataFrame, should have 'image_png' column specifying the image name
        destination_folder: Path to the folder where you want images to copy to
        source_folder: Path to the folder where images are currently at
        
    Returns:
        Images will be copied to the destination_folder from the source_folder
    """
    
    
    # Make sure destination exists
    os.makedirs(destination_folder, exist_ok=True)

    # Loop over all image filenames in your dataframe
    for filename in data["image_png"]:
        source_path = os.path.join(source_folder, filename)
        dest_path = os.path.join(destination_folder, filename)
        
        # Copy file if it exists
        if os.path.exists(source_path):
            shutil.copy2(source_path, dest_path)
        else:
            print(f"⚠️ File not found: {source_path}")
```

```python
# Dictionary: {Category_Name: [1,5,7]}
top_8_gr1 = {
    'Obsidian':[5],
    'Pegmatite':[10, 16],
    'Breccia':[11],
    'Chert': [10],
    'Conglomerate':[5],
    'Rock Gypsum':[4],
    'Rock Salt':[1]   
}

top_8_gr2 = {
    'Obsidian': [6],
    'Pegmatite':[3, 11],
    'Peridotite':[13],
    'Rock Salt': [13, 15],
    'Chert': [13],
    'Conglomerate':[14]
}
```

```python
# For easier access
df = ave_dist_rat_df.drop(['Group1','Group2'], axis=1).copy()
df
```

```python
# select top images
df_top_1 = df[ df.apply(
    lambda row: row['Item'] in top_8_gr1.get(row['Category_Name'], []), 
    axis=1) ]

df_top_2 = df[ df.apply(
    lambda row: row['Item'] in top_8_gr2.get(row['Category_Name'], []), 
    axis=1) ]
```

```python
# Compare top distinctivenesses accros groups
stats.ttest_ind(df_top_1['Distinct_Rating'], df_top_2['Distinct_Rating'])
```

```python
# Create a dictionary for bottom images
bottom_items = {
    'Andesite': [3, 4, 6, 8, 9, 10],
    'Basalt': [2, 9, 1, 8, 13, 10],
    'Diorite': [10, 1, 4, 15, 5, 3],
    'Gabbro': [2, 8, 11, 12, 14, 16],
    'Rhyolite': [5, 8, 10, 11, 14, 16],
    'Phyllite': [5, 3, 7, 10, 13, 2],
    'Quartzite': [12, 7, 8, 15, 5, 13],
    'Slate': [3, 4, 10, 11, 14, 15],
    'Dolomite': [1, 14, 16, 7, 13, 12],
    'Sandstone': [8, 4, 11, 3, 16, 15],
    'Pumice': [8, 11, 12, 13, 1, 14],
    'Gneiss': [5, 6, 10, 15, 13, 14],
    'Marble': [3, 4, 16, 2, 14, 6],
    'Shale': [1, 2, 3, 7, 8, 12]
}

# Randomly shuffle it
random.seed(1)
for key in bottom_items:
    random.shuffle(bottom_items[key])

# first 3 go to group 1, last 3 - to group 2
bottom_42_gr1 = {key: values[:3] for key, values in bottom_items.items()}
bottom_42_gr2 = {key: values[3:] for key, values in bottom_items.items()}
```

```python
# Select bottom images
df_bottom_1 = df[df.apply(
    lambda row: row['Item'] in bottom_42_gr1.get(row['Category_Name'], []),
    axis = 1
)]
df_bottom_2 = df[df.apply(
    lambda row: row['Item'] in bottom_42_gr2.get(row['Category_Name'], []),
    axis = 1
)]
```

```python
# Compare dist ratings
stats.ttest_ind(df_bottom_1['Distinct_Rating'], df_bottom_2['Distinct_Rating'])
```

```python
# Add distinctiveness type, group, and concatenate them
df_top_1 = df_top_1.copy()
df_top_2 = df_top_2.copy()
df_bottom_1 = df_bottom_1.copy()
df_bottom_2 = df_bottom_2.copy()

# Distinct_Type 
df_top_1.loc[:,'Distinct_Type'] = 'Top'
df_top_2.loc[:,'Distinct_Type'] = 'Top'

df_bottom_1.loc[:, 'Distinct_Type'] = 'Bottom'
df_bottom_2.loc[:, 'Distinct_Type'] = 'Bottom'

# Combine by groups
df_1 = pd.concat([df_top_1,df_bottom_1],ignore_index=True)
df_2 = pd.concat([df_top_2,df_bottom_2],ignore_index=True)

# Add groups
df_1.loc[:, 'Group1'] = 'Target'
df_1.loc[:, 'Group2'] = 'Foil'

df_2.loc[:, 'Group1'] = 'Foil'
df_2.loc[:, 'Group2'] = 'Target'

full_df = pd.concat([df_1, df_2], ignore_index=True)
full_df
```

```python
# Compare the whole data set dist ratings between groups
stats.ttest_ind(
    full_df.loc[full_df['Group1'] == 'Target', 'Distinct_Rating'],
    full_df.loc[full_df['Group2'] == 'Target', 'Distinct_Rating']
)
```
