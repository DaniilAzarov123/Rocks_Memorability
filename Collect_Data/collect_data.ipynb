# Both conditions

```python
# Load the data from the CSV files
df_by_item_2cond = pd.read_csv('../Exp3_Conds_A_B/Exp3_Analysis/csv files/df_by_item.csv')
filtered_data_memory_2cond = pd.read_csv('../Exp3_Conds_A_B/Exp3_Analysis/csv files/filtered_data_memory.csv')
filtered_data_test_2cond = pd.read_csv('../Exp3_Conds_A_B/Exp3_Analysis/csv files/filtered_data_test.csv')
```

## Memory & Test phases

```python
# Unique sorted IDs
unique_ids = set(filtered_data_test_2cond['subject_id'])
unique_ids = list(unique_ids)
unique_ids.sort()

# new IDs
new_ids = range(1, len(unique_ids) + 1)
ids_dict = dict(zip(unique_ids, new_ids))
filtered_data_memory_2cond['subject_id'] = filtered_data_memory_2cond['subject_id'].map(ids_dict)
filtered_data_test_2cond['subject_id'] = filtered_data_test_2cond['subject_id'].map(ids_dict)
```

```python
def re_encode_dfs(df_by_item, filtered_data_memory, filtered_data_test, conditions_1_3_present=True):
    
    # ========== filtered_data_memory & filtered_data_test encoding ==========
    
    # Trial index
    filtered_data_test['trial_index'] = (
        filtered_data_test
        .groupby('subject_id')
        .cumcount() + 1
    )
    filtered_data_memory['trial_index'] = (
        filtered_data_memory
        .groupby('subject_id')
        .cumcount() + 1
    )
    
    # Group encoding
    filtered_data_test['group'] = np.where(filtered_data_test['group'] == 'Group1', 1, 2)
    filtered_data_memory['group'] = np.where(filtered_data_memory['group'] == 'Group1', 1, 2)
    
    # Condition encoding
    # 0 - 1
    # 1 - 3
    if conditions_1_3_present:
        filtered_data_test['condition'] = np.where(filtered_data_test['condition'] == '1_Rep', 0, 1)
        filtered_data_memory['condition'] = np.where(filtered_data_memory['condition'] == '1_Rep', 0, 1)
    
    # Item type encoding
    # 0 - Foil
    # 1 - Target
    filtered_data_test['item_type'] = np.where(filtered_data_test['item_type'] == 'Target', 1, 0)
    filtered_data_memory['item_type'] = np.where(filtered_data_memory['item_type'] == 'Target', 1, 0)
    
    # Response encoding
    # 0 - a - No - not seen before
    # 1 - l - Yes - seen before
    filtered_data_test['response'] = np.where(filtered_data_test['response'] == 'a', 0, 1)
    filtered_data_test['correct_response'] = np.where(filtered_data_test['correct_response'] == 'a', 0, 1)
    filtered_data_test['cor_ans'] = np.where(filtered_data_test['cor_ans'] == False, 0, 1)
    
    # Distinctiveness type encoding
    # 0 - Low Distinctiveness
    # 1 - High Distinctiveness
    filtered_data_test['distinct_type'] = np.where(filtered_data_test['distinct_type'] == 'Top', 1, 0)
    filtered_data_memory['distinct_type'] = np.where(filtered_data_memory['distinct_type'] == 'Top', 1, 0)
    
    # "Is a good subject?" encoding
    # 0 - False -Bad subject
    # 1 - True - Good subject
    filtered_data_test['good_subj'] = np.where(filtered_data_test['good_subj'] == True, 1, 0)
    filtered_data_memory['good_subj'] = np.where(filtered_data_memory['good_subj'] == True, 1, 0)
    
    # Task encoding
    # 0 - Memory phase
    # 1 - Test phase
    filtered_data_test['task'] = 1
    filtered_data_memory['task'] = 0
    
    # ========== df_by_item encoding ==========
    
    # Condition encoding
    if conditions_1_3_present:
        df_by_item['condition'] = np.where(df_by_item['condition'] == '1_Rep', 0, 1)

    # Group encoding
    df_by_item['group'] = np.where(df_by_item['group'] == 'Group1', 1, 2)

    # Item type encoding
    # 0 - Foil
    # 1 - Target
    df_by_item['item_type'] = np.where(df_by_item['item_type'] == 'Foil', 0, 1)

    # Distinctiveness type encoding
    # 0 - Low Distinctiveness
    # 1 - High Distinctiveness
    df_by_item['distinct_type'] = np.where(df_by_item['distinct_type'] == 'Top', 1, 0)

    # Response type encoding
    # 0 - False Alarm
    # 1 - Hit
    df_by_item['resp_type'] = np.where(df_by_item['resp_type'] == 'Hit', 1, 0)
    
    return df_by_item, filtered_data_memory, filtered_data_test
```

```python
# Re-encode the dataframes
df_by_item_2cond, filtered_data_memory_2cond, filtered_data_test_2cond = re_encode_dfs(df_by_item_2cond, filtered_data_memory_2cond, filtered_data_test_2cond, conditions_1_3_present=True)
```

```python
# Drop str columns
df_by_item_2cond.drop(columns=['category_name'], inplace=True)
filtered_data_memory_2cond.drop(columns=['category_name','broad_category','image_png'], inplace=True)
filtered_data_test_2cond.drop(columns=['category_name','broad_category','image_png'], inplace=True)
```
